```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diamond_df = pd.read_csv("https://raw.githubusercontent.com/ktxdev/AIM-5001/main/M8/1.%20Data/diamonds.csv")

# Helper functions
def univariate_analysis_categorical(data, var, order = None):
    '''
        Plots a barplots for categorical variables using both Matplotlib and Seaborn 
    '''
    value_counts = data[var].value_counts()
    
    if order:
        value_counts = value_counts.reindex(order)

    fig, ax = plt.subplots(1, 2, figsize=(20, 8))

    # Bar plot in Matplotlib
    ax[0].bar(value_counts.index, value_counts.values)
    ax[0].set_title(f"Freq Dist of {var} Indicator (Matplotlib)")

    # Count plot in Seaborn
    ax[1].set_title(f"Freq Dist of {var} Indicator (Seaborn)")
    sns.barplot(x = value_counts.index, y = value_counts.values, ax = ax[1])

    fig.tight_layout()
    
def univariate_analysis_continuous(data, var):
    '''
        Plots boxplots and histograms for continuous variables using both Matplotlib and Seaborn 
    '''
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 8))

    ax1.boxplot(data[var], labels=[""])
    ax1.set_title(f'Boxplot for {var} (Matplotlib)')

    ax2.hist(data[var], bins = 10)
    ax2.set_title(f'Histogram for {var} (Matplotlib)')

    sns.boxplot(data[var],  orient = 'v', ax = ax3)
    ax3.set_title(f'Boxplot for {var} (Seaborn)')

    sns.histplot(data[var], bins = 10, ax = ax4)
    ax4.set_title(f'Histogram for {var} (Seaborn)')

    fig.tight_layout()
```

## Part 3f: Comparing boxplots of carat metric for each distinct clarity value

```python
# Group data by the 'clarity' column
grouped_data = diamond_df.groupby('clarity')

# Create an empty list to store data for each clarity level
boxplot_data = []

# Iterate over each group
for group_name, group_df in grouped_data:
    boxplot_data.append(group_df['carat'])

# Plot boxplots
fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(20, 8))

ax1.boxplot(boxplot_data, labels = grouped_data.groups.keys())
ax1.set_title(f'Boxplot of carat by clarity level (Matplotlib)')

sns.boxplot(x ='clarity', y = 'carat', data = diamond_df, ax = ax2)
ax2.set_title(f'Boxplot of carat by clarity level (Seaborn)')

fig.tight_layout()
```

```python
# Calculating range and IQR for all clarity values
for clarity_level in diamond_df['clarity'].unique():
    clarity_level_df = diamond_df[diamond_df['clarity'] == clarity_level]
    # Calculate range
    clarity_range = clarity_level_df['carat'].max() - clarity_level_df['carat'].min()

    # Calculate IQR
    Q1 = clarity_level_df['carat'].quantile(0.25)
    Q3 = clarity_level_df['carat'].quantile(0.75)
    IQR = Q3 - Q1
    
    # Median
    median = clarity_level_df['carat'].median()
    
    print(f"{clarity_level}: Range = {clarity_range}; IQR = {IQR}; median = {median}")
```

The box plots show that diamonds with clarity 'I1' have the biggest variation in terms of weight (i.e. carat) at 4.71, while clarity 'VS2' has the greatest Interquatile variation at 0.64, implying that clarity 'VS2' has a little higher weight variability in the middle of the data set. 

## 3g: Scatter plot of carat vs. depth

Plotting carat (i.e. weight) vs depth reveals that the majority of diamonds in the dataset have a weight of roughly 56 to 62 and a depth of approximately 0.2 to 2.5. Furthermore, the plot demonstrates that the association between carat and depth remains constant.
